```python
import math

def base(x):
    num = ""
    letra = 1
    for i in range(len(x)):
        if(x[i] == "K"):
            if((i == (len(x)-2) and x[i+1] != "B") or (i != (len(x)-1) and (i != (len(x)-2)))):
                raise Exception
            else: letra = 2**10
            break
        elif(x[i] == "M"):
            if((i == (len(x)-2) and x[i+1] != "B") or (i != (len(x)-1) and (i != (len(x)-2)))):
                raise Exception
            else: letra = 2**20
            break
        elif(x[i] == "G"):
            if((i == (len(x)-2) and x[i+1] != "B") or (i != (len(x)-1) and (i != (len(x)-2)))):
                raise Exception
            else: letra = 2**30
            break
        elif(x[i] == "B"):
            if(i != (len(x)-1)):
                raise Exception
            break
        else: num = num + x[i]
    
    return (int(num))*letra

def baseinv(x):
    if((x%(2**30)) == 0):
        return str(int(x/(2**30)))+"G"
    elif((x%(2**20)) == 0):
        return str(int(x/(2**20)))+"M"
    elif((x%(2**10)) == 0):
        return str(int(x/(2**10)))+"K"
    else:
        return str(int(x))

def hexbin(h):
    b = ""
    
    for i in range(len(h)):
        if h[i] == "0": b += "0000"
        elif h[i] == "1": b += "0001"
        elif h[i] == "2": b += "0010"
        elif h[i] == "3": b += "0011"
        elif h[i] == "4": b += "0100"
        elif h[i] == "5": b += "0101"
        elif h[i] == "6": b += "0110"
        elif h[i] == "7": b += "0111"
        elif h[i] == "8": b += "1000"
        elif h[i] == "9": b += "1001"
        elif h[i] == "A": b += "1010"
        elif h[i] == "B": b += "1011"
        elif h[i] == "C": b += "1100"
        elif h[i] == "D": b += "1101"
        elif h[i] == "E": b += "1110"
        elif h[i] == "F": b += "1111"
        else: raise Exception
        
    print("\nConversão\nHexadecimal ---> Binário\n"+h+" ---> "+b)
    return b

while True:
    print("\nMemória Cache com Mapeamento Associativo Completo\n"+
          "[1] Cálculo da quantidade de bits necessários para uma determinada memória cache\n"+
          "[2] Cálculo de formato de endereço para memórias cache com mapeamento associativo completo\n"+
          "[3] Determinar o valor binário do campo bloco que será localizado pelo sistema de controle da cache\n")

    while True:
        try:
            escolha = int(input("Escolha uma das opções: "))
            if(escolha < 1 or escolha > 3):
                raise Exception()
            break
        except (ValueError, Exception):
            print("Valor inválido!")

    if(escolha == 1):
        print("\nCálculo da quantidade de bits necessários para uma determinada memória cache")

        while True:
            try:
                Ncache = base(input("Informe a capacidade da memória cache: "))
                break
            except Exception:
                print("Valor inválido!")

        while True:
            try:
                X = base(input("Informe a largura das linhas da cache (células): "))
                break
            except Exception:
                print("Valor inválido!")

        while True:
            try:
                Np = base(input("Informe a capacidade de células da MP: "))
                break
            except Exception:
                print("Valor inválido!")  
            
        #Total de bits de dados
        Tdados = Ncache*8
        #Quantidade de linhas
        L = Ncache/X
        #Quantidade de blocos
        B = Np/X
        #Largura do campo bloco
        bloco = int(math.log2(B))
        #Total de bits do campo bloco
        Tbloco = L*bloco
        #Total de bits da cache
        Tcache = Tdados + Tbloco

        print("\nTotal de bits necessários para a memória cache:\nT = ",baseinv(Tcache),"bits")
        
    if(escolha == 2):
        print("\nCálculo de formato de endereço para memórias cache com mapeamento associativo completo")

        while True:
            try:
                Np = base(input("Informe a capacidade de células da MP: "))
                break
            except Exception:
                print("Valor inválido!")  
        
        while True:
            try:
               L = base(input("Informe o número de linhas da cache: "))
               break
            except Exception:
                print("Valor inválido!")

        while True:
            try:
                X = base(input("Informe a largura da linha da cache (Bytes): "))
                break
            except Exception:
                print("Valor inválido!")

        #Total de Blocos
        Tblocos = Np/X
        #Largura do campo Bloco
        bloco = int(math.log2(Tblocos))
        #Largura do campo Byte (Célula)
        celula = int(math.log2(X))
        #Endereço
        print("\nFormato de endereço da cache:\n| Bloco |",bloco," bits\n| Byte  |",celula," bits\nTotal de bits do endereço: ",(bloco+celula)," bits\n"
              +"Número de linhas da cache: ",baseinv(L))

    if(escolha == 3):
        print("\nDeterminar o valor binário do campo bloco que será localizado pelo sistema de controle da cache")
        
        while True:
            try:
                X = base(input("Informe a largura dos blocos (Bytes) da MP: "))
                break
            except Exception:
                print("Valor inválido!")

        while True:
            try:
                Ncache = base(input("Informe a capacidade da cache (número de células): "))
                break
            except Exception:
                print("Valor inválido!")

        while True:
            try:
                A = input("Informe o endereço de acesso em hexadecimal: ")
                Abin = hexbin(A)
                break
            except Exception:
                print("Valor inválido!") 

        #Bits do campo Célula (Byte)
        celula = int(math.log2(X))
        #Quantidade de blocos
        B = (2**(len(A)*4))/X
        #Bits do campo Bloco
        bloco = int(math.log2(B))
        #Endereço em binário
        Abloco = ""
        Acelula = ""
        for i in range(len(Abin)):
            if(i>=0 and i<bloco):
                Abloco = Abloco + Abin[i]
            elif(i>=bloco):
                Acelula = Acelula + Abin[i]

        print("\nFormato de endereço da cache:\n| Bloco |",Abloco,"\n| Byte  |",Acelula,"\nEndereço do bloco para acesso: ", Abloco)
        
    print("\n[1] Sim\n[0] Não")
    while True:
        try:
            prosseguir = int(input("Realizar outra operação? "))
            if(prosseguir != 0 and prosseguir != 1):
                raise Exception()
            break
        except (ValueError, Exception):
            print("Valor inválido!")

    if(prosseguir == 0):
        print("Programa encerrado")
        break
```
